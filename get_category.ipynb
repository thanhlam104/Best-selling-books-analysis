In this file, we extract book links from the scraped general data and pass to amazon_category spider to scrape the category from the detail page of each book.
After finish the scraping process, we merge the new data with the old one.

```python
import pandas as pd
import numpy as np
file = pd.read_json('amazon_raw.json')
file['link'] = file.book_href.apply(lambda x: '/'.join(x.split('/')[0:5])+'/'+x.split('/')[-1].split('?')[0] if x else None)
file['dp'] = file.book_href.apply(lambda x: x.split('/')[-1].split('?')[0] if x else None)
file.drop(['author_href', 'book_href'], 1, inplace = True)
file
```

```python
file_dp = file[['link','dp']].drop_duplicates('dp')
file_dp.to_csv('links.csv')
```

```python
cate = pd.read_csv('scrapy\cate1.csv')
cate['dp'] = cate['ASIN']
cate = cate.drop('ASIN',1)
cate = cate.dropna().drop_duplicates()
```

```python
with_cate = pd.merge(file, cate, on='dp', how = 'left').drop_duplicates()
with_cate.drop(['link','dp'],1, inplace=True)
```

```python
with_cate.to_csv('amazon.csv')
```
